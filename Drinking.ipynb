```python
le = pd.read_csv('F:/Pycharm/Data Analytics/Drinking/lifeexpectancy-verbose.csv')

keep_col = ['GhoDisplay', 'RegionDisplay', 'WorldBankIncomeGroupDisplay', 'CountryDisplay', 'SexDisplay',
            'YearDisplay', 'DisplayValue']
le2 = le[keep_col]
le2.to_csv('lifeexpectancy.csv')
le2
```

```python
le2.head()
```

```python
le2.tail()
```

```python
le2 = le2.groupby(['CountryDisplay'],as_index=False)['DisplayValue'].sum()
le2 = le2.sort_values('DisplayValue', ascending=False)
le2.head()
```

```python
le2 = le2[:]
plt.figure(figsize=(40, 40))
sns.set_style('darkgrid')
plt.xticks(rotation=45, ha='right', fontsize=11)
plt.xlabel('Name of Country', fontsize=30)
plt.ylabel('Life Expectancy Age', fontsize=30)
p = sns.barplot(x='CountryDisplay', y='DisplayValue', data=le2)
p.set(xlabel='Name of Country', ylabel='Life Expectancy Age')
p.set_title('Life Expectancy Per Country')
```

```python
le3 = le.groupby(['WorldBankIncomeGroupDisplay'], as_index=False)['DisplayValue'].sum()
le3 = le3.sort_values('DisplayValue', ascending=False)
le3.head()
```

```python
le3 = le3[:]
plt.figure(figsize=(40,40))
plt.xticks(rotation=45, ha='right', fontsize=40)
p = sns.barplot(x='WorldBankIncomeGroupDisplay', y='DisplayValue', data=le3)
p.set(xlabel='Region Area', ylabel='Life Expectancy Age')
p.set_title('Life Expectancy Per Region')
```
